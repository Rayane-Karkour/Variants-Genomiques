```python
# Used to determine the final KEEP_FIELDS
def column_summary(df, order_by=2, ascending=True):
    summary = []
    for col in df.columns:
        n_unique = df[col].nunique()
        n_null = df[col].isnull().sum()
        pct_null = (n_null / len(df) * 100).round(1) #How much percent is null 

        non_null_values = df[col].dropna() # Remove null so I can get an sample of the data
        example = str(non_null_values.iloc[0])
        if len(example)> 30:
            example = example[:30]+"..." 
        summary.append((col, n_unique, n_null, pct_null, example))
    
    summary_df = pd.DataFrame(summary, columns=['Column', 'Unique', 'Nulls', 'Percent_Null', 'Value'])
    
    if order_by == 0 : # Unique
        summary_df = summary_df.sort_values('Column', ascending=ascending)
    elif order_by == 1: # Null
        summary_df = summary_df.sort_values('Unique', ascending=ascending)
    elif order_by == 2: # Alphabetical
        summary_df = summary_df.sort_values('Nulls', ascending=ascending)
    else:
        raise ValueError('''Order_by must be 
        - 0 : alphabetical
        - 1 : unique
        - 2 : nulls
        ''')

    print(f'   Columns   |     Unique (w/o NA)   |        nulls         |    Value')
    print("-"*70)
    for _, row in summary_df.iterrows():
        print(f'{row["Column"]:<12} | {row["Unique"]:>6} uniques values | {row["Nulls"]:>5} nulls ({row["Percent_Null"]:<5}%) | {str(row["Value"]):<20}')
```

## Keeping some columns
Alright, not all of these columns contain important information, some have too many missing values, while others provide redundant details.
So we will only select the columns that include the information necessary to describe, interpret, and analyze genetic variants :

---
- CHROM : chromosome where the variant is located (31 values, 0% missing)
- POS : genomic position of the variant
- REF : reference allele in the reference genome
- ALT : alternate allele observed in the variant

**Why ?** : Well, these four fields define the genomic variant, the 4 of them represent the minimum information required to locate and describe a mutation

---
- ID : variant identifier from the dataset.
- ALLELEID : unique ClinVar allele identifier.
- RS : dbSNP reference SNP identifier.

**Why ?** : ID always make it easier to reference and search in datasets !  
As for ALLELEID and RS, they are also useful in case I want to link the dataset with another dataset

---
- CLNSIG : clinical significance.
- CLNREVSTAT : review status of the clinical interpretation.
- CLNDN : disease names associated with the variant.

**Why ?** : First : Human readable data, Second : Gives important description about the relevance of the variant and its diseases

---
- GENEINFO : genes associated with the variant.
- MC : molecular consequence.
- CLNVC : variant type.
- ORIGIN : indicates whether the variant is germline, somatic, or unknown.

**Why ?** : Similar to the list before , information about the impact of the variant, mutation type and consequences

```python
KEEP_FIELDS = [
    'CHROM',
    'POS',
    'ID',
    'REF',
    'ALT',
    'CLNVC',
    'ALLELEID',
    'GENEINFO',
    'CLNSIG',
    'CLNREVSTAT',
    'CLNDN',
    'MC',
    'ORIGIN',
    'RS',
]

df = df[KEEP_FIELDS]
```

## Signification (CLNSIG)
*Explanation from the doc*

### **CLNSIG**
Aggregate germline classification for this single variant;   
Multiple values are separated by a vertical bar

In humain terme : The clinical classification of this variant (for example: benign, pathogenic).

```python
# Distribution
clnsig_counts = df['CLNSIG'].value_counts(dropna=False)
print(f'Number of different value (including NA): {len(clnsig_counts)}')
print(clnsig_counts) #.to_string() to show full
```

```python
# The cooler Distribution
top_n = 11 # Cutting at 'other' (around > 1000) or else it looks like nothing
top_cat = clnsig_counts.head(top_n)
top_cat['Other'] = clnsig_counts.iloc[top_n:].sum()

top_cat.plot(kind='bar')
plt.ylabel('Count')
plt.xlabel('CLNSIG')
plt.title('Distribution of CLNSIG values')
plt.show()
```

## Distribution per chromosome

```python
#Needed fpr orderding from 1–22 , else it orders like strings (1,11,2,22,...)
def chrom_key(c):
    if c.isdigit():
        return int(c) 
    else:
        return 23

print(sorted(df['CHROM'].unique(), key=chrom_key))

# After checking out what those NT/NW (Not chromosome) I've settled on making the list myself
CHROM_ORDER = [str(i) for i in range(1, 23)] + ['X', 'Y']
```

```python
# Compute counts for each category
all_counts = df['CHROM'].value_counts().reindex(CHROM_ORDER, fill_value=0)
unknown_counts = df[df['CLNSIG'] == 'Uncertain_significance']['CHROM'].value_counts().reindex(CHROM_ORDER, fill_value=0)
likely_benign_counts = df[df['CLNSIG'] == 'Likely_benign']['CHROM'].value_counts().reindex(CHROM_ORDER, fill_value=0)
benign_counts = df[df['CLNSIG'] == 'Benign']['CHROM'].value_counts().reindex(CHROM_ORDER, fill_value=0)
path_counts = df[df['CLNSIG'] == 'Pathogenic']['CHROM'].value_counts().reindex(CHROM_ORDER, fill_value=0)
```

```python
plt.bar(CHROM_ORDER, all_counts.values, color='lightgray', label='All variants')
plt.bar(CHROM_ORDER, unknown_counts.values, label='Uncertain')
plt.bar(CHROM_ORDER, likely_benign_counts.values, bottom=unknown_counts.values, label='Likely benign')
plt.bar(CHROM_ORDER, benign_counts.values, bottom=unknown_counts.values + likely_benign_counts.values, label='Benign')
plt.bar(CHROM_ORDER, path_counts.values, bottom=unknown_counts.values + likely_benign_counts.values + benign_counts.values, label='Pathogenic')

plt.ylabel('Variants')
plt.title('Variants per chromosome')
plt.legend()

plt.show()
```

## Gene analysis

```python
# Exctract all Gene name form geneinfo
# GENEINFO > OR4F5:79501
df['GENE'] = df['GENEINFO'].str.split(':').str[0]
```

```python
print("Most frequent genes (limited to 20): ")
print(df['GENE'].value_counts().head(20))
```

```python
# Top genes for all and pathogenic variants
top_all  = df['GENE'].value_counts().head(20)
top_path = df[df['CLNSIG'] == 'Pathogenic']['GENE'].value_counts().reindex(top_all.index, fill_value=0)

x = np.arange(len(top_all))  # positions for genes
width = 0.4  # bar width

plt.figure(figsize=(14, 6))
plt.bar(x - width/2, top_all.values, width, label='All variants')
plt.bar(x + width/2, top_path.values, width, label='Pathogenic variants')

plt.xticks(x, top_all.index, rotation=90)
plt.ylabel('Number of Variants')
plt.title('Top Genes : All vs Pathogenic Variants')
plt.legend()
plt.show()
```

```python
# SNV vs Indels par CLNSIG
df['VARIANT_CLASS'] = df['IS_SNV'].map({True: 'SNV', False: 'Indel/Other'})

top5_sigs = df['CLNSIG'].value_counts().head(5).index
df_plot = df[df['CLNSIG'].isin(top5_sigs)]

ct = pd.crosstab(df_plot['CLNSIG'], df_plot['VARIANT_CLASS'], normalize='index') * 100

fig, ax = plt.subplots(figsize=(10, 5))
ct.plot(kind='bar', stacked=True, ax=ax,
        color=['#3b82f6', '#f97316'], edgecolor='white', linewidth=0.5)
ax.set_ylabel('% de variants')
ax.set_title('Proportion SNV vs Indels par Signification Clinique', fontsize=12)
ax.legend(title='Type', bbox_to_anchor=(1.01, 1))
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

## Pathogenic
The problem with searching for pathogenes is that the “Pathogenic” data is split across multiple possible values, such as "Likely_pathogenic" or "Pathogenic/Pathogenic".
To handle this, I will separate them into two categories: 
- Pure Pathogenic : Only with the tag Pathogenic since it's the most popular
- Pathogenic Likely : Probably a pathogene , but may include other columns  

So with this we can try to analyze pathogenes.

```python
# Pure "Pathogenic" only
pure_path_df = df[df['CLNSIG'] == 'Pathogenic']

# 'contains' allows to check like a regex so it get all Pathogenics rows
path_likely_df = df[df['CLNSIG'].str.contains('Pathogenic', na=False)]
```

```python
# Stats for pure Pathogenic variants
pure_pathogenic_stats = pd.DataFrame({
    'Metrics': [
        'Total variants',
        'Unique genes',
        'Gene with the most variants',
        'Total pathogenic variants',
        'Unique genes with variants',
        '% of total variants=pathogenic',
    ],
    'Values': [
        f"{len(df):,}",
        f"{df['GENE'].nunique():,}",
        f"{pure_path_df['GENE'].value_counts().idxmax()} ({pure_path_df['GENE'].value_counts().max():,} variants)",
        f"{len(pure_path_df):,}",
        f"{pure_path_df['GENE'].nunique():,}",
        f"{len(pure_path_df)/len(df)*100:.2f}%",
    ]
})
print('Pure')
pure_pathogenic_stats
```
